```python
df = df.drop_duplicates()
```

```python
# create needed columns 
df["Revenue"] = df["Price"] * df["Reviews"]
```

```python
df["Price Category"] = pd.cut(df["Price"],
                             bins=[0,10,20,50,100],
                             labels=["Cheap","Mid","High","Luxury"])
```

```python
# popularity score (simple version)
df["Popularity"] = df["User Rating"] * df["Reviews"]
```

```python
# sort by reviews from the most popular
df.sort_values(by="Reviews", ascending =False).reset_index(drop=True)
```

```python
df.groupby("Genre")["Reviews"].sum()
```

```python
df.groupby("Author")["Reviews"].sum()
```

```python
df.groupby(["Author","User Rating"])["Reviews"].mean().reset_index().sort_values(by="User Rating", ascending=False)
```

```python
# using aggregations to pass math functions

df.groupby(["Author","User Rating"])["Reviews"].agg(["mean","max","median"]).reset_index().sort_values(by="max", ascending=False)
```
